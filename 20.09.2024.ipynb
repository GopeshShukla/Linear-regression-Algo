```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('student_info.csv')
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
#data visualization
plt.scatter(x=df.study_hours,y=df.student_marks)
plt.title("Student Data")
plt.xlabel("Study Hours")
plt.ylabel("Study Hours")
plt.show()
```

```python
# finding null values
df.isnull()

```

```python
# finding null values 
df.isnull().sum()
```

```python
#finding avg
df.mean()
```

```python
#putting average of a partparticular culumn on the place of nan of that column.

df2=df.fillna(df.mean())
```

```python
df2
```

```python
# separate dependent and independent data
x=df2.drop("student_marks",axis=1)
y=df2.drop("study_hours",axis=1)
x.head()
```

```python
#splitting dataset into traing and testing

from sklearn.model_selection import  train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
# print(x_train.size,x_test.size,y_train.size,y_test.size,x.size,y.size)
```

```python
from sklearn.linear_model import LinearRegression
#       |         |                   |
#      \/        \/                  \/
#    library  module               class
lr = LinearRegression()  #object (lr) is called model.
lr.fit(x_train,y_train)

```

```python
result=lr.predict([[8]])[0][0].round(3)
output = float(result)
print(output)
```

```python
result=lr.predict([[6]])[0][0].round(3)
output = float(result)
print(output)
```

```python
result=lr.predict([[600]])[0][0].round(3)
output = float(result)
print(output)
```

```python
from sklearn.metrics import r2_score

y_pred = lr.predict(x_test)
# Using the trained model (lr) to make predictions on the test data (x_test).

accuracy = round(r2_score(y_test, y_pred) * 100, 2)
# Using the built-in round() function to round the accuracy to two decimal places.

accuracy
# Displaying the final accuracy of the model.
```
